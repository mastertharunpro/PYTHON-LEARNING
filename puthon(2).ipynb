IDENTITY OPERATORS: [IS,IS NOT]

- IT IS USED TO FIND THE VSRIABLE OF THE TWO GIVEN STRINGS 
-  IT CHECKS ID VALUE ARE SAME (OR) NOT

```python
a = 23
b = 23

br = a is b 
print(br)
br = a is not b 
print(br)
```
