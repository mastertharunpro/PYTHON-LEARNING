# FUNCTIONS

- SET OF CODE THAT CAN BE REUSED

```python
def addition(a,b,c):
    return a+b+c


addition(b=5,a=6,c=7)
```

# WITH PARAMETER: 

- THE GIVEN DATA INSIDE THE CODE

```python
def addition(a, b):
    a = 10
    b = 10
    return a + b

print(addition(1, 2))
```

# WITHOUT PARAMETER 
- THE GIVEN DATA OUTSIDE OF CODE

```python
def addition(a,b,c):
    return a+b+c


addition(b=5,a=6,c=7)
```
